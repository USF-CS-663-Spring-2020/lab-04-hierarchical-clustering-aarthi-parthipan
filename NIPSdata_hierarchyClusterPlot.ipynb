```python
import pandas as pd

df = pd.read_csv("NIPS_1987-2015.csv")

df
```

```python
dfFeatures = df.iloc[:,1:]
dfFeatures
```

```python
# While the implementation in sklearn is good,
# the scipy version can produce the dendrogram
from scipy.cluster.hierarchy import linkage, dendrogram

Z = linkage(dfFeatures, 'ward')
```

```python
# Draw the dendrogram
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
dendrogram(Z)
plt.show()
```
