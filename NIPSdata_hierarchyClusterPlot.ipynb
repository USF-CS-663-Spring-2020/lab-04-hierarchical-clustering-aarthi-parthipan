```python
import pandas as pd

df = pd.read_csv("NIPS_1987-2015.csv")

df
```

```python
dfFeatures = df.iloc[:,1:]
dfFeatures
```

```python
df.rename(columns = { "Unnamed: 0" :"list of words"} , inplace = True)
df
```

```python
dfLabels = df.iloc[:,0:1]
dfLabels
```

```python
df['word usage count']= df.iloc[:, 1:].sum(axis=1)
df
```

```python
df2 = df.set_index('list of words')
```

```python
df2
```

```python
# While the implementation in sklearn is good,
# the scipy version can produce the dendrogram
from scipy.cluster.hierarchy import linkage, dendrogram

Z = linkage(df2, 'ward')
```

```python
# Draw the dendrogram
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure(figsize=(30,20))
dendrogram(Z)
#dendrogram(Z, orientation="right", labels = df2.index, leaf_font_size=8, show_contracted=True)
plt.xlabel('distance')
plt.show()
```

```python
#picking the top 100 records and removing the last column of usage count
df3 = df2.sort_values(by='word usage count', ascending = False).head(100).iloc[:,:5812]
df3
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

Z2 = linkage(df3, 'ward')
```

```python
plt.figure(figsize=(30,50))
#dendrogram(Z2)
dendrogram(Z2, 
           orientation="right",  
           labels = df3.index,
           leaf_font_size=10, 
           show_contracted=True)
plt.xlabel('distance')
plt.show()
```
